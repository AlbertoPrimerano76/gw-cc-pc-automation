```python
import xml.etree.ElementTree as ET

model_space = "{http://guidewire.com/datamodel}"

policy_center = "/Users/aprimerano/Development/NZ/PolicyCenter"
policy_center_config = policy_center + "/modules/configuration/config"
policy_center_entities = policy_center_config + "/extensions/entity"

lob_file = policy_center_entities + "/NLLLiabilityLine.eti"

tree = ET.parse(lob_file)
root = tree.getroot()


# This are the columns and the typekey the only fields that needs to be passed
# Find all elements with the tag "namespace + column" and meet specific criteria
matching_elements = [elem for elem in root.iter() if
                     ( elem.tag == model_space + "column" or
                       elem.tag == model_space + "typekey"
                     ) and
                     not elem.get("name", "").endswith("Avl") and
                     not elem.get("getterScriptability") == "doesNotExist"]
```

```python
xml_to_print = ""

type_mapping = { 'bit' : ['xmlns:xs="http://www.w3.org/2001/XMLSchema"' , 'type="xs:boolean"'] ,
                 'datetime' : ['xmlns:ns1="http://guidewire.com/xsd"', 'gw:type="java.util.Date"'],
                 'integer':['xmlns:xs="http://www.w3.org/2001/XMLSchema"' , 'type="xs:int"'],
                 'shorttext': ['xmlns:xs="http://www.w3.org/2001/XMLSchema"' , 'type="xs:string"'],  
                 'decimal': ['xmlns:xs="http://www.w3.org/2001/XMLSchema"' , 'type="xs:decimal"'],
                 'longtext': ['xmlns:xs="http://www.w3.org/2001/XMLSchema"' , 'type="xs:string"']  
                }
unmateched_elements = []

for elem in matching_elements:
    if elem.get("typelist") is not None:
        type = 'type="xs:string"'
        xmls = 'xmlns:xs="http://www.w3.org/2001/XMLSchema"'
    else:
        value = type_mapping.get(elem.get("type"))
        xmls = value[0]
        type = value[1]
        
    xml_to_print += '<xs:attribute {} name="{}" use="optional" {}"/> \n'.format(xmls, elem.get("name"),type)
    #else:
   #     unmateched_elements.append(elem.get("name"))
    
```
